```python
custom_data_file = open("Custom_Data.txt", "r")
custom_data = custom_data_file.readlines()

custom_dataset = []
data_point = None
for line in custom_data:
  if line[0] == "#":
    if data_point:
      data_point['code'] = ''.join(data_point['code'])
      custom_dataset.append(data_point)
    data_point = {"text": None, "code": []}
    data_point['text'] = line[1:]
  else:
    data_point["code"].append(line)

import pandas as pd
custom_dataset = pd.DataFrame(custom_dataset)
custom_dataset = custom_dataset[custom_dataset['text'].str.contains("python" or "Python")]
```

```python
from IPython.display import display, HTML
display(HTML(custom_dataset.to_html()))
```
