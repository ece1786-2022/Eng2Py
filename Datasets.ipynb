#### Training Data

```python
training_data_file = open("Training_after_mods.txt", "r")
training_data = training_data_file.readlines()
```

```python
training_data[0] = training_data[0][training_data[0].find('#'):]
```

```python
training_dataset = []
data_point = None
count = 0
for line in training_data[:-1]:
  if line[0] == "#":
    if data_point:
      data_point['code'] = ''.join(data_point['code'])
      training_dataset.append(data_point)
    data_point = {"text": None, "code": []}
    data_point['text'] = line[1:]
  else:
    data_point["code"].append(line)
```

```python
import pandas as pd
training_dataset = pd.DataFrame(training_dataset)
training_dataset['text'] = training_dataset['text'].str.strip()
```

```python
from IPython.display import display, HTML
display(HTML(training_dataset.to_html()))
```

```python
from pathlib import Path
filepath = Path('/content/drive/MyDrive/ECE1786_Project/Training.csv')
training_dataset.to_csv(filepath, index = False)
```
